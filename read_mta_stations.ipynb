```python
def read_mta_stations():
    ''' Read the MTA master list into a pd.DataFrame (Put the CSV file on the repository!!)
    data source = https://data.cityofnewyork.us/Transportation/Subway-Stations/arq3-7z49 (export data as csv)
    '''
    df = pd.read_csv("mta_stations.csv")
    df.rename(columns = {"OBJECTID": "key", "NAME":"name", "the_geom":"location"}, inplace = True)
    
    df["lon"] = df.location.str.split().str.get(1).str.strip('(')
    df["lat"] = df.location.str.split().str.get(2).str.strip(')')
    df = df[["name","LINE","lon","lat"]]
    df = df.sort_values(["name","LINE"], ascending = [True, True]).reset_index()
    
    return df
    
    '''
    return pd.DataFrame (columns = [station (unique), lattitude, longitude])
    '''
```

```python
companies['key'] = 0
stations['key'] = 0
cs = companies.merge(stations, how='left', on = 'key')
cs.drop('key',1, inplace=True)
```

```python
cs.sample(0)
```

```python
from geopy.distance import geodesic as gd
```

```python
def dist_calc(row):
    start = (row['LAT'], row['LON'])
    end = (row['lat'], row['lon'])
    return gd(start,end).miles

cs['distance'] = cs.apply(lambda row: dist_calc(row), axis =1)
cs
```

```python
min_d = (
            cs.
            sort_values(by =['COMPANY', 'distance'], ascending = [True, True]).
            groupby('COMPANY').
            head(5).
            reset_index()
        )
```

```python
min_d.rename(columns={'name':'stations'}, inplace = True)
min_d
```
